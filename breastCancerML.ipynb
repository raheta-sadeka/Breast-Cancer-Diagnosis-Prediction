# Breast Cancer Diagnosis Prediction using K-Nearest Neighbors (KNN) and K-Means Clustering



## Import Libraries

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler
from sklearn.neighbors import KNeighborsClassifier
from sklearn.cluster import KMeans
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, classification_report
```

## Load Dataset

```python
df = pd.read_csv("Dataset.csv")   # change file name if needed
df.head()
```

## Data Preprocessing

```python
df = df.drop(["id", "Unnamed: 32"], axis=1)
```

## Encode diagnosis: M → 1, B → 0

```python
df["diagnosis"] = df["diagnosis"].map({"M": 1, "B": 0})
```

## Split into features & target

```python
X = df.drop("diagnosis", axis=1)
y = df["diagnosis"]
```

## Apply Min-Max Normalization

```python
scaler = MinMaxScaler()
X_scaled = scaler.fit_transform(X)
```

## Train-Test Split

```python
X_train, X_test, y_train, y_test = train_test_split(
    X_scaled, y, test_size=0.2, random_state=42
)
```

## K-Means Clustering (k=2)

```python
kmeans = KMeans(n_clusters=2, random_state=42)
clusters = kmeans.fit_predict(X_scaled)
```

## Add cluster results to original dataframe for observation

```python
df["cluster"] = clusters
df[["diagnosis", "cluster"]].head(15)
```

## KNN Classifier (k=5)

```python
knn = KNeighborsClassifier(n_neighbors=5)
knn.fit(X_train, y_train)

y_pred = knn.predict(X_test)
```

## Evaluation Metrics

```python
accuracy = accuracy_score(y_test, y_pred)
precision = precision_score(y_test, y_pred)
recall = recall_score(y_test, y_pred)
f1 = f1_score(y_test, y_pred)

print("KNN Model Evaluation:\n")
print(f"Accuracy  : {accuracy:.4f}")
print(f"Precision : {precision:.4f}")
print(f"Recall    : {recall:.4f}")
print(f"F1-Score  : {f1:.4f}")

print("\nClassification Report:\n")
print(classification_report(y_test, y_pred))
```
